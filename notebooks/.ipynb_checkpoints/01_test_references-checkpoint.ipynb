# 01 — Test references

Checks own dense ED and optional NetKet references.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.home() / "nes_lattice_project"
# If this project lives somewhere else, change PROJECT_ROOT manually.
sys.path.insert(0, str(PROJECT_ROOT / "src"))

from nes_lattice.hamiltonians import make_hamiltonian_spec
from nes_lattice.references import get_reference_energies
```

```python
# Small 1D TFIM: own ED should work even without NetKet
hspec = make_hamiltonian_spec("tfim", shape=(6,), g=1.0, pbc=True)
energies, source = get_reference_energies(hspec, k=4, prefer="auto", own_ed_max_sites=14)
print(source)
print(energies)
```

```python
# Small 2D TFIM: NetKet if installed, otherwise own ED if small enough
hspec = make_hamiltonian_spec("tfim", shape=(3,3), g=1.0, pbc=True)
energies, source = get_reference_energies(hspec, k=3, prefer="auto", own_ed_max_sites=14)
print(source)
print(energies)
```

```python
# 2D Heisenberg in Sz=0 sector
hspec = make_hamiltonian_spec("heisenberg", shape=(4,2), J=1.0, pbc=True)
energies, source = get_reference_energies(hspec, k=3, prefer="auto", own_ed_max_sites=14)
print(source)
print(energies)
```
