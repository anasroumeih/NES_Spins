# 04 — Compare ansätze

Run the same system with FFN, CNN, and RBM. This can take time.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.home() / "nes_lattice_project"
sys.path.insert(0, str(PROJECT_ROOT / "src"))

from nes_lattice.train import TrainConfig, train, save_history
```

```python
base = dict(
    shape=(4,4),
    hamiltonian="tfim",
    k=2,
    g=1.0,
    steps=800,
    lr=1e-3,
    n_chains=64,
    n_samples=8,
    print_every=200,
)

for model in ["ffn", "cnn", "rbm"]:
    cfg = TrainConfig(**base, model=model, seed={"ffn":0, "cnn":1, "rbm":2}[model])
    params, history = train(cfg)
    path = PROJECT_ROOT / "results" / f"compare_2d_tfim_4x4_k2_{model}.json"
    save_history(history, path, cfg)
    print("saved", path)
```
